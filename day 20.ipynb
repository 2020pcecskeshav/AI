```python
def logic(row):
    profit_ratio=(row['earning']-row['cost'])/row['duration']
    if profit_ratio>=3:
        return row['status']=='blockbuster'
    elif profit_ratio>=2:
        return row['status']=='superhit'    
    elif profit_ratio>=1:
        return row['status']=='hit' 
    elif profit_ratio>=0:
        return row['status']=='average'
    else :
        return row['status']=='flop'      

movie['validation']=movie.apply(logic,axis=1) 
movie.head()    
```
